```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

```python
credit_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_data.head()
```

```python
credit_data.tail()
```

```python
credit_data.info()
```

```python
credit_data['Class'].value_counts()
```

```python
# seperate the data for analysis
legit = credit_data[credit_data.Class==0]
fraud = credit_data[credit_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the value of both the transactions
credit_data.groupby('Class').mean()
```

```python
legit_sample=legit.sample(n=222)
```

```python
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
x = new_data.drop(columns='Class' ,axis=0)
y = new_data['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
# Split the data
X_train,X_test,Y_train,Y_test=train_test_split(x,y ,test_size=0.2, stratify=y, random_state= 2)
```

```python
print(X_train.shape,X_test.shape,x.shape)
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

Accuracy score

```python
X_train_prediction= model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction= model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(test_data_accuracy)
```
